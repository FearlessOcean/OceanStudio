```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data//raw//online_shopping_10_cats.csv',usecols=['label','review'],encoding='utf-8').dropna()
```

```python
df.head
```

```python
# 划分数据集  注意分层抽样
train_df,test_df = train_test_split(df,test_size=0.2,stratify=df['label'])
```

```python
from src.tokenizer import JiebaTokenizer
```

```python
#计算序列长度 （为了pad填充做准备）
tokenizer = JiebaTokenizer.from_vocab('data//vocab.txt')
train_df['review'] = train_df['review'].apply(lambda x : len(tokenizer.tokenize(x))).quantile(0.95)
```
